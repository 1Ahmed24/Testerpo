```python
df['Cumulative Distance'] = df['Total Distance'].cumsum()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Start Time', y='Cumulative Distance', data=df)
plt.title('Cumulative Distance Traveled Over Time')
plt.xlabel('Start Time')
plt.ylabel('Cumulative Distance (km)')
plt.show()
"This line plot shows the cumulative distance traveled by all vehicles over time. It provides a clear picture of the total mileage covered and helps identify trends or patterns in vehicle usage over different periods."
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans



# Select features for clustering
features = ['Total Distance', 'Fuel Consumption', 'Max Speed', 'Idle Running']

# Convert selected features to numeric and handle non-numeric values
for feature in features:
    df[feature] = pd.to_numeric(df[feature], errors='coerce')

# Drop rows with missing values in the selected features
df_clean = df.dropna(subset=features)

# Check if the cleaned dataframe is empty
if df_clean.empty:
    print("No data available for clustering after cleaning. Please check the dataset.")
else:
    # Standardize features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(df_clean[features])

    # Apply KMeans clustering
    kmeans = KMeans(n_clusters=3, random_state=0).fit(scaled_features)
    df_clean['Cluster'] = kmeans.labels_

    # Plot the clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Total Distance', y='Fuel Consumption', hue='Cluster', data=df_clean, palette='viridis')
    plt.title('Clusters of Trips')
    plt.xlabel('Total Distance (km)')
    plt.ylabel('Fuel Consumption (liters)')
    plt.show()
```
